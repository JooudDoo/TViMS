```python
import numpy as np
from matplotlib import pyplot

def upper(p):
    return (1-p)/(p**3+1-p)

def footer(p):
    return ( (1-p)**2 + (1-p)**2 * p ) / ( p**2 + (1-p)**2 + p * (1-p)**2)

def uppfotby2(p,coef):
    return (upper(p)+footer(p))*coef

with open("file.txt") as f:
    line = f.readline()
    res = line.split()
    res=  list(map(float, res))

with open("file1.txt") as f:
    line = f.readline()
    x = line.split()
    x = list(map(float, x))

uppersY = np.apply_along_axis(upper, 0, x)
footersY = np.apply_along_axis(footer, 0, x)

pyplot.plot(x, res, color="green")
#pyplot.plot(x, uppersY, color="yellow")
#pyplot.plot(x, footersY, color="red")
for i in range(1, 10000):
    for m in range(1,10000):
        upfot = np.apply_along_axis(uppfotby2, 0, x, coef=m/i)
        iscor = True
        for val in range(len(upfot)):
            if abs(upfot[val] - res[val]) >= 0.02:
                iscor = False
                break
        if iscor:
            pyplot.plot(x, upfot)
            print(i, m)
```
